```python
import pandas as pd
import yaml
dir = '/Users/Liu/NBA_Pro_Line_Analytics/'

with open(dir + 'config.yaml', 'r') as stream:
    try:
        config = yaml.load(stream)
    except yaml.YAMLError as exc:
        print(exc)
```

```python
pd.set_option('display.max_columns', 70)
```

```python
df_new = pd.read_excel(dir + 'data/NBA_Team_Stats_2010-2019/{0}'.format(config['NBA_Team_stats'][0]), sheet_name= 0)
```

```python
df_new = df_new.replace('New Jersey', 'Brooklyn')
```

```python
df_new = df_new.drop(['MAIN REF', 'CREW', 'HALFTIME','BOX SCORE\nURL','ODDS\nURL','OPENING ODDS', 
                      'LINE \nMOVEMENT #1', 'LINE \nMOVEMENT #2','LINE \nMOVEMENT #3','TO',
                      'CLOSING\nODDS'], axis=1)
```

```python
df_new = df_new.rename(columns = {'STARTING LINEUPS': 'SF', 
                                  'Unnamed: 38': 'SF2',
                                  'Unnamed: 39':'C',
                                  'Unnamed: 40': 'SG2',
                                  'Unnamed: 41': 'SG1',
                                  'TO\nTO':'Total_TO',
                                  'TEAM\nREST DAYS': 'Num_Rest_Days',
                                  'F':'Final_Score'})
```

```python
df_new = df_new.set_index(['GAME-ID'])
```

```python
df_new['FT_PCT'] = df_new['FT']/df_new['FTA']
df_new['FG_PCT'] = df_new['FG']/df_new['FGA']
df_new['3P_PCT'] = df_new['3P']/df_new['3PA']
df_new['TS_PCT'] = df_new['PTS']/(2*(df_new['FGA']+.44*df_new['FTA']))
df_new['DATE'] = pd.to_datetime(df_new['DATE'])
```

```python
df_new
```

```python
#AVG PPG last 5 games
poww = pd.DataFrame()
for team in df_new.TEAM.unique():
    aa = df_new[df_new.TEAM == team].groupby('TEAM')['PTS'].rolling(5).mean().shift().reset_index()
    poww = poww.append(aa)
poww = poww.rename(columns = {'PTS':  'PPG_AVG_L5G'})
df_new = df_new.reset_index()
df_new = df_new.merge(poww, how='left', on=['TEAM', 'GAME-ID'])
df_new = df_new.set_index(['GAME-ID'])
```

```python
#Pts scored last game
poww = pd.DataFrame()
for team in df_new.TEAM.unique():
    aa = df_new[df_new.TEAM == team].groupby('TEAM')['PTS'].rolling(1).mean().shift().reset_index()
    poww = poww.append(aa)
df_new = df_new.reset_index()
poww = poww.rename(columns={'PTS': 'PTS_LG'}) 
df_new = df_new.merge(poww, how='left', on=['TEAM', 'GAME-ID'])
df_new = df_new.set_index(['GAME-ID'])
```

```python
#AVG PPG last 3 games
poww = pd.DataFrame()
for team in df_new.TEAM.unique():
    aa = df_new[df_new.TEAM == team].groupby('TEAM')['PTS'].rolling(3).mean().shift().reset_index()
    poww = poww.append(aa)
poww = poww.rename(columns={'PTS': 'PPG_AVG_L3G'})
df_new = df_new.reset_index()
df_new = df_new.merge(poww, how='left', on=['TEAM', 'GAME-ID'])
df_new = df_new.set_index(['GAME-ID'])
```
