```python
from hyperopt import fmin, STATUS_OK, Trials, hp, tpe, rand
import lightgbm
import pandas as pd 
import numpy as np
from model_build_scripts.NBA_model_build import NBA_Model_Build
```

```python
#Defines the model scenario in ./configs/scenarios_config.yml
scenario_name = 'pro_line_model_build_all_params'
```

```python
scenario = NBA_Model_Build(scenario_name)
```

```python
scenario.load_data()
```

```python
initial_params = {'objective': 'multiclass',
                  'metric': 'multi_logloss',
                  'num_boost_round': 1000,
                  'early_stopping_rounds': 20,
                  'num_classes': 3,
                  'learning_rate': .05,
                  'min_data_in_leaf': 25,
                  'seed': scenario.seed
                 }
```

```python
#Runs Recursive Feature Elimination to Remove all features of 0 Importance
post_rfe_model, var_importance  = scenario.run_rfe(model_params= initial_params, target = scenario.target, X_vars= scenario.X_vars)
```

```python
#Runs Hyperparameter tuning (Bayesian Optimization) - current scenario set to complete 200 rounds 
best_params, hyperparam_obj = scenario.run_hyperopt(scenario.hyperopt_param_space, list(post_rfe_model.booster_.feature_name()), 
                                                    model_params = initial_params, fmin_max_evals = scenario.fmin_max_evals)
```

```python
best_params
```

```python
static_params =  {'objective': 'multiclass',
               'metric': 'multi_logloss',
               'num_boost_round': 1000,
               'early_stopping_rounds': 20,
               'num_classes': 3,
               'seed': scenario.seed
            }
```

```python
best_params['num_leaves'] = int(best_params['num_leaves'])
best_params['min_data_in_leaf'] = int(best_params['min_data_in_leaf'])
```

```python
final_model = lightgbm.LGBMModel(**best_params, **static_params, importance_type = 'gain')
eval_set = [(scenario.df_tune[list(post_rfe_model.booster_.feature_name())], scenario.df_tune[scenario.target])
```

```python
final_model.fit(X =scenario.df_train[list(post_rfe_model.booster_.feature_name())],
                y = scenario.df_train[scenario.target],
                eval_set = eval_set)
```
